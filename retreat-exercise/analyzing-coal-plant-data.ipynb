# now import the coal plants spreadsheet

NOTE `pandas` is a Python package, and `read_excel()` is a method/function within that package; using this `.function()` notation is called object-oriented programming; in simple terms, `pandas.read_excel(arguments)` creates an object below called `excel_file`, which we can then do other stuff to

you'll see this kind of syntax a lot when coding in Python

# now do some basic manipulations of the coal dataframe

## let's look at the summed plant capacity in MW by status

```python
sum_capacity_dataframe = pandas.DataFrame(coal_units_dataframe.groupby('Status')['Capacity (MW)'].sum())
```

```python
sum_capacity_dataframe
```

### sort it by decreasing capacity

```python
sum_capacity_dataframe.sort_values('Capacity (MW)', ascending=False, inplace=True)
```

### create a quick bar plot using `.plot.bar()`

```python
sum_capacity_dataframe.plot.bar()
```

## create a table with region as rows and summed capacity across columns

```python
capacity_dataframe = pandas.DataFrame(coal_units_dataframe.groupby(['Region','Status'])['Capacity (MW)'].sum()).unstack()
capacity_dataframe.columns = capacity_dataframe.columns.droplevel()
```
